```python
#dropping_table
df=df.drop('Sex',axis=1)
df
```

```python
#heatmap
import matplotlib.pyplot as plt
import seaborn as sns
corr=df.corr()
sns.heatmap(corr,fmt='.2g',annot=True,linewidth=1)
plt.tight_layout()
plt.show()
```

```python
#Elbow_method
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",random_state=40)
  kmeans.fit(df)
  wcss.append(kmeans.inertia_)
plt.figure(figsize=(10,5))
plt.plot(range(1,11),wcss)
plt.title('elbow method')
plt.xlabel('n_clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#Kmeans_clustering
kmeans=KMeans(n_clusters=3,init="k-means++",random_state=40)
y_kmeans=kmeans.fit_predict(df)
```

```python
y_kmeans
```

```python
df1=df.values
```

```python
#visualizing_the_clusters
plt.figure(figsize=(15,7))
plt.scatter(df1[y_kmeans==0,6],df1[y_kmeans==0,7],color='grey',label='cluster1',s=50)
plt.scatter(df1[y_kmeans==1,6],df1[y_kmeans==1,7],color='pink',label='cluster2',s=50)
plt.scatter(df1[y_kmeans==2,6],df1[y_kmeans==2,7],color='skyblue',label='cluster3',s=50)
plt.scatter(kmeans.cluster_centers_[:,6],kmeans.cluster_centers_[:,7],color='red',label='Centroids',s=150,marker='d')
plt.grid(False)
plt.title('Clusters')
plt.xlabel('Shucked weight')
plt.ylabel('Viscera weight')
plt.legend()
plt.show()
```

```python
#add_column_clusters
df['Clusters']=pd.DataFrame(y_kmeans)
df
```
